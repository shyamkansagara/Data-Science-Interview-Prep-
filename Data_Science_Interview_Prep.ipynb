

##From a list of numbers, find the squares of the numbers using the map function 




```python
def sq(x):
  return x*x

n1 = [1,2,3,5,6,68,8,8]
squared_elements = map(sq,n1)

for i in squared_elements:
  print(i)
```

## Filter the numbers divisible by 3 from a tuple using lambda function 

```python
tup = (1,3,5,6,7,8,9,23,55,6,67)
```

```python
a = list(filter((lambda x:x%3 == 0 ),tup)) # on filter we can create any type of object that we want 
for i in a:
  print(i)
```

## Difference betn iloc() and loc()

```python
# iloc = index location 
a = { 'a':[1,2,3,6,7,8] , 'b' : [2,1,0,5,6,7],'c' : [2,1,0,5,6,7]}
import pandas as pd

a1 = pd.DataFrame(a) #here column a should have more than 1 value in the values of dict. As dataframe cannot take scalar value. You can try it and can see the error
a1
```

```python
a1.loc[:5,['a']] # if we see, 5 the element in the row in not exclueded. For the iloc()
```

```python
a1.iloc[:5,:3] # this wont work with the loc()
```

```python
a1.loc[a1['a']>3,['b','c']]# Whenever we use the Column names for any conditions just dont use the iloc(. Use it only when we are using the indexes)
```

```python
# Lets use iloc AND TRY THAT OUT
a1.iloc[(a1['a']>3).values,1:3]
```

##Difference between list.append() and list.extend()

```python
l1 = ['In' , 'out']

l1.append([1,2]) # to add one element to the list 
print(l1)
l1.extend([1,2]) # to extend the list. Basically, concat two list 
print(l1)
```
